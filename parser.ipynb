```python
def stockfish_evaluation(board, time_limit = 0.02):
    engine = chess.engine.SimpleEngine.popen_uci("stockfish_14.1_win_x64_avx2\stockfish_14.1_win_x64_avx2.exe")
    result = engine.analyse(board, chess.engine.Limit(time=time_limit,depth=20))
    print(result)
    res = result["score"]
    if res.is_mate():
        res = res.white().score(mate_score=10)
    else:
        res = res.relative.score()
    return res
score = [] 
for f in FEN:
    board = chess.Board(f)
    print(board)
    result = stockfish_evaluation(board)
    score.append(result)
print(score)
```
